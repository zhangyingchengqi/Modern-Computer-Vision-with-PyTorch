## 第二章 第四节 张量与ndarray的对比

```python
import torch
x = torch.rand(1, 6400)
y = torch.rand(6400, 5000)
print( x.shape, y.shape )
```

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
assert device == 'cuda', "This exercise assumes the notebook is on a GPU machine"
```

```python
x, y = x.to(device), y.to(device) #将创建的张量注册到设备中。 意味着在设备中存储信息
```

```python
# 1.在GPU上运算的时间
# 单行计时魔法命令（对应的多行版本是 %%timeit）。
#会多次运行你后面的那条语句，然后取平均耗时，帮助评估代码性能
%timeit z=(x@y)
```

```python
# 2.在CPU上执行相同的运算
x, y = x.cpu(), y.cpu()
%timeit z=(x@y)
```

```python
# 3. 在numpy数组上执行
import numpy as np
x = np.random.random((1, 6400))
y = np.random.random((6400, 5000))
%timeit z = np.matmul(x,y)
```
